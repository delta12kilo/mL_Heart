```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
scale_columns = ['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_sodium']
df[scale_columns] = StandardScaler.fit_transform(df[scale_columns])
```

```python
df.head()
```

```python
X = df.drop(['DEATH_EVENT'],axis=1)
y = df['DEATH_EVENT']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
print('x-test: ',X_test.size)
print('xtrain: ',X_train.size)
print('ytrain: ',y_train.size,' ytest: ',y_test.size)
```

### Logistic Regration

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model1 = lr.fit(X_train,y_train)
prediction1 = model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,prediction1)
```

```python
cm
```

```python
sns.heatmap(cm,annot=True,cmap='BuPu')
```

```python
tp = cm[0][0]
tn = cm[1][1]
fn = cm[1][0]
fp = cm[0][1]
print('test acc: ',(tp+tn)/(tp+tn+fn+fp))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction1))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
model2 = dtc.fit(X_train,y_train)
prediction2 = model2.predict(X_test)
cm2 = confusion_matrix(y_test,prediction2)
```

```python
cm2
```

```python
accuracy_score(y_test,prediction2)
```

```python
print(classification_report(y_test,prediction2))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
model3 = rfc.fit(X_train,y_train)
prediction3 = model3.predict(X_test)
cm3 = confusion_matrix(y_test,prediction3)
```

```python
cm3
rfca = accuracy_score(y_test,prediction3)
print('Rrfc scoree: ',rfca)
```

```python
print(classification_report(y_test,prediction3))
```
